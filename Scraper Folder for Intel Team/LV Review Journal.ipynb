###Las Vegas Review journal scrape

```python
"""
Description:

This script scrapes the Las Vegas Review Journal website for news stories, and then saves the information to a .csv file.
Content extracted includes:
- Source name
- Article URL
- Date and Time
- Article Headline/Title
- Article Content
"""
```

```python
# Load the relevant packages:
from bs4 import BeautifulSoup
import requests
import numpy
import gettext
import re
import csv
from itertools import zip_longest
```

```python
# Specify news site:
site = requests.get('https://www.reviewjournal.com/local/')

# Get site content:
soup = BeautifulSoup(site.content, 'html5lib')
```

```python
# Create an empty list for url, date/time, headline, content:
urls = []
date_time = []
headlines = []
contents = []

number_of_articles = 50

try:
  
  for n in numpy.arange(0, number_of_articles):

# Find all urls. Append them to urls list:
    for news in soup.findAll('div',{'entry-title field-title'}):
        urls.append(news.a['href'])
# Remove duplicate hyperlinks:
        urls = list(dict.fromkeys(urls))
    
# Extracting the news articles:
    for link in urls:
        article = requests.get(link)
        page_content = article.content
        soup_content = BeautifulSoup(page_content, 'html5lib')
        body = soup_content.find_all('p')
    
# Extracting titles:
        title = soup_content.find_all('h1', class_='entry-title')
        titles = title[n].get_text()
        headlines.append(titles)

# Extracting dates:
        date = soup_content.find_all('div', class_='rj-field-published-time')
        dates = date[n].get_text()
        temp_date = []
        temp_date.append(dates)
        for item in temp_date:
            item = re.sub('\t+','',item)
            item = re.sub('\n','',item)
            date_time.append(item)

# Unifying the paragraphs
        list_paragraphs = []
        for p in numpy.arange(0, len(body)):
            paragraph = body[p].get_text()
            list_paragraphs.append(paragraph)
            final_article = " ".join(list_paragraphs)
        
        contents.append(final_article)
except IndexError:
  pass
```

```python
sources = ['las_vegas_review'] * len(urls) # Creates list comprising entirely of "las_vegas_review", with length of the other lists

# Save scraped information to a labeled csv file:
scraped_data = [sources, urls, date_time, headlines, contents]
#export_data = zip_longest(*scraped_data, fillvalue = '')
#with open('las_vegas_review.csv', 'w', encoding="UTF-8", newline='') as file:
      #write = csv.writer(file)
      #write.writerow(("source","url","publishedDateTime","headline","content"))
      #write.writerows(export_data)
```
